### 1. Setup - Importar librerías

```python
import pandas as pd
```

### 2. Configuración

```python
data_input = "../data/processed/germany.veterinarians.csv"
data_output = "../data/processed/germany.dashboard.csv"

columnas_categoricas = ["status of employment", "field of work", "federal state", "community size", "gender", "income"]

columnas_ERI = ["ERI_1","ERI_2","ERI_3","ERI_4","ERI_5","ERI_6","ERI_7","ERI_8","ERI_9","ERI_10"]
columnas_CBI = ["CBI_01","CBI_02","CBI_03","CBI_04","CBI_05","CBI_06"]
columnas_OTHERS = ["REDONDEAR","gender","status of employment","field of work","federal state","community size","income"]
```

### 3. Mapeo de valores

```python
status_of_employment = {
    1: "Self-Employed",
    2: "Employed",
    3: "Unemployed by Choice",
    4: "Currently looking for a job",
    5: "Parental Leave",
    6: "Working in a non-veterinarian field",
    7: "Retired",
    8: "Medical Leave",
    9: "Other" # Translated for consistency
}

field_of_work = {
    1: "Small Animal Medicine",
    2: "Horses",
    3: "Farm Animals",
    4: "Mixed practice",
    5: "Food Industry",
    6: "Slaughterhouse",
    7: "R&D",
    8: "Industry Technical Service",
    9: "Industry Field Service",
    10: "Public Health",
    11: "Research",
    12: "University",
    13: "Zoo Animals",
    14: "Other",
    15: "Other"
}

federal_state = {
    1: "Baden-Wurttemberg",
    2: "Bavaria",
    3: "Berlin",
    4: "Brandenburg",
    5: "Bremen",
    6: "Hamburg",
    7: "Hessen",
    8: "Mecklenburg-Vorpommern",
    9: "Niedersachsen",
    10: "North Rhine-Westphalia",
    11: "Rheinland-Pfalz",
    12: "Saarland",
    13: "Saxony (Sachsen)",
    14: "Saxony-Anhalt",
    15: "Schleswig-Holstein",
    16: "Thuringia"
}

gender = {
    1: "Male",
    2: "Female"
}

community = {
    1: "5000 pop. aprox.",
    2: "20000 pop. aprox.",
    3: "50000 pop. aprox.",
    4: "100000 pop. aprox."
}

income = {
    1: "1. less than 1000 eur",
    2: "2. 1000 to 2000 eur",
    3: "3. 2000 to 3000 eur",
    4: "4. 3000 to 4000 eur",
    5: "5. 4000 to 5000 eur",
    6: "6. more than 5000 eur"
}

working_hours = {
    1: "0-20",
    2: "21-40",
    3: "41-60",
    4: "61-80",
    5: "81 or more"
}

eri_scale = {
    1: "Strongly disagree",
    2: "Disagree",
    3: "Agree",
    4: "Strongly agree"
}

cbi_scale = {
    100: "Always",
    75: "Often",
    50: "Sometimes",
    25: "Seldom",
    0: "Never"
}

cbi_score = {
    1: 100,
    2: 75,
    3: 50,
    4: 25,
    5: 0
}
```

### 4. Cargar los datos

```python
# Cargar datos
print("-" * 50)
print("Cargando datos...")
print("-" * 50)

df = pd.read_csv(data_input)
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
print("-" * 50)
print("\nPrimeras 5 filas:")
display(df.head())
```

### 5. Transformacion de datos para el dashboard

```python
dashboard = df.copy()
dashboard["status of employment"] = dashboard["status of employment"].map(status_of_employment)
dashboard["field of work"] = dashboard["field of work"].map(field_of_work)
dashboard["federal state"] = dashboard["federal state"].map(federal_state)
dashboard["gender"] = dashboard["gender"].map(gender)
dashboard["community size"] = dashboard["community size"].map(community)
dashboard["income"] = dashboard["income"].map(income)
dashboard[columnas_ERI] = dashboard[columnas_ERI].replace(eri_scale)
dashboard[columnas_CBI] = dashboard[columnas_CBI].replace(cbi_scale)
```
